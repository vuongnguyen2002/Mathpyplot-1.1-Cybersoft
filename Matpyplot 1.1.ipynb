```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(3,5)
y= 2*x+1
plt.plot(x ,y ,color='r')
plt.plot(x, y, '-r', label='y=2x+1')
plt.title('y=2x+1')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')
plt.show()
```

```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-5,5)
y= 2*(x**2)+3*x+4
plt.plot(x ,y ,color='r')
plt.plot(x, y, '-r', label='y=2*(x**2)+3*x+4')
plt.title('y=2*(x**2)+3*x+4')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')
plt.show()
```
